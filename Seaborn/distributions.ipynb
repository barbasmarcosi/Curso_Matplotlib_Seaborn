```python
import numpy as np
import seaborn as sns
import scipy
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=50, hue='smoker', stat='density', multiple='fill')
```

```python
sns.histplot(data=tips, x='tip', bins=50, cumulative=True)
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', fill=True)
```

```python
sns.ecdfplot(data=tips, x='tip', hue='smoker', stat='count')
```

```python
sns.displot(data = tips, x='tip', hue='sex', kind='hist', multiple='stack')
```
