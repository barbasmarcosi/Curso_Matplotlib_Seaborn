```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr());
```

```python
sns.heatmap(tips.corr(), annot=True);
```

```python
sns.heatmap(data=tips[['total_bill','tip']], annot=True);
```

```python
sns.heatmap(data=tips.corr(), annot=True, cmap='coolwarm_r',
            linewidths=5, linecolor='black', vmin=0.49, vmax=1, cbar=False)
```
