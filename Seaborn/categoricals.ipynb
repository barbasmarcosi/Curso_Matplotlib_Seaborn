```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True,color='black')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```python
sns.catplot(tips, x='day', y='total_bill', hue='smoker', dodge=True, kind='swarm', col='time');
```
