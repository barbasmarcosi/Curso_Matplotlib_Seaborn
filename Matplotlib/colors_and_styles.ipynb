```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

```python
fix, axes = plt.subplots(figsize=(8,8))
axes.plot(x, x**2, 'ro:')
axes.plot(x, x**3, 'b:')
axes.plot(x, x**4, 'y--')
axes.plot(x, x**5, 'g.-')
```

```python
fix, axes = plt.subplots(figsize=(8, 8))
axes.plot(x, x**2, color='green', alpha=0.8, linewidth=3, marker='o', markersize=10, markerfacecolor='white')
axes.plot(x, x**3, color='blue',  alpha=0.2, linestyle='--', marker='x')
axes.plot(x, x**4, color='#66f4a2', linestyle='dashed', marker='v')
axes.plot(x, x**5, color='salmon', marker='P')
```
