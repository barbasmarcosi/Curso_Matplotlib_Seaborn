```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'USA', 'GERMANY']
population = np.random.rand(6) * 10e12
population
```

```python
plt.bar(country, population, width=0.5, color=['red', 'blue'])
plt.xticks(np.arange(6), ('INDIA', 'JAPAN', 'MEXICO',
           'COLOMBIA', 'USA', 'GERMANY'), rotation=45)
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
